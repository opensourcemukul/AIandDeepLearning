```python
from keras.preprocessing.text import Tokenizer
```

```python
import nltk
from nltk.tokenize import word_tokenize
import numpy as np
import re
```

```python
from doc3 import training_doc3
```

```python
cleaned = re.sub(r'\W+', ' ', training_doc3).lower()
```

```python
cleaned
```

```python
nltk.download('punkt')
```

```python
tokens = word_tokenize(cleaned)
print(tokens)
```

```python
text_sequences = []
```

```python
train_len = 4 #3(input)+1(output)
```

```python
for i in range(train_len, len(tokens)):
  seq = tokens[i-train_len: i]
  text_sequences.append(seq)
```

```python
text_sequences
```

```python
sequences = {}
count = 1
for i in range(len(tokens)):
  if tokens[i] not in sequences:
    sequences[tokens[i]] = count
    count = count+1
```

```python
sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
tokenizer
```

```python
sequences
```

```python
vocabulary_size = len(tokenizer.word_counts) + 1
```

```python
vocabulary_size
```

```python
n_sequences = np.empty([len(sequences), train_len], dtype = 'int32')
```

```python
n_sequences
```

```python
for i in range(len(sequences)):
  n_sequences[i] = sequences[i]
```

```python
n_sequences
```

```python
train_inputs = n_sequences[:,:-1]
train_target = n_sequences[:,-1]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
train_target = to_categorical(train_target, num_classes = vocabulary_size)
```

```python
seq_len = train_inputs.shape[1]
```

```python
train_inputs.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
```

```python
model = Sequential()
model.add(Embedding(vocabulary_size, seq_len, input_length = seq_len))
model.add(LSTM(50, return_sequences = True))
model.add(LSTM(50))
model.add(Dense(25, activation='relu'))
model.add(Dense(vocabulary_size, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_inputs, train_target, epochs = 300, verbose=1)
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
input_text = input().strip().lower()

encoded_text = tokenizer.texts_to_sequences([input_text])[0]

pad_encoded = pad_sequences([encoded_text], maxlen = seq_len, truncating ='pre')

print(encoded_text, pad_encoded)
```

```python
for i in (model.predict(pad_encoded)[0]).argsort()[-3:][::-1]:
  pred_word = tokenizer.index_word[i]
  print("next word: ",pred_word)
```
