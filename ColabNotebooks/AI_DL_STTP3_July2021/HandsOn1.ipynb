```python
# to make a simple calculator
# want to make a function: 3 ips-> operand1, operand2, operator
def calculator(operand1, operand2, operator):
  if operator == '+':
    return operand1 + operand2
  elif operator == '-':
    return operand1 - operand2
  elif operator == '*':
    return operand1 * operand2
  elif operator == '/':
    #need to check if operand2 is 0 or not
    if operand2 != 0:
      return operand1 / operand2
    else:
      return None
  else:
    print("invalid input")
    return None
```

```python
print(calculator(100, 10, '/'))
```
