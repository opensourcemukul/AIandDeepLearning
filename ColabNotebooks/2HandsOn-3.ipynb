#Linear Regression

```
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
X = np.array([1,6,9]) #input
Y = np.array([1,2,3]) #output
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(X,Y)
```

```
model = LinearRegression()
```

```
X_train = X.reshape(-1,1)
```

```
X_train
```

```
X
```

```
Y_train = Y
```

```
model.fit(X_train,Y_train)
```

```
Y_predicted = model.predict(X_train)
```

```
Y_predicted
```

```
plt.scatter(X,Y)
plt.plot(X_train, Y_predicted, color='red')
plt.show()
```

```
model.coef_
```

```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv("car_dataset.csv")
```

```
data
```

```
data = data.drop("Unnamed: 0", axis=1)
```

```
data
```

```
data = data.drop("Id", axis=1)
```

```
data
```

```
data = data.drop("Vin", axis=1)
```

```
data
```

```
data['Make'] = data['Make']+data['Model']
```

```
data
```

```
data = data.drop("Model", axis=1)
```

```
data
```

```
data['City'] = data['City'].str.lower()
data['City'] = data['City'].str.replace(" ","")
data['State'] = data['State'].str.lower()
data['Make'] = data['Make'].str.lower()
data['Make'] = data['Make'].str.replace(" ","")
data['Make'] = data['Make'].str.replace("-","")
```

```
data
```

```
data.info()
```

```
data_bkp = data
```

```
data=pd.get_dummies(data)
```

```
data
```

```
data.describe()
```

```
X = data
```

```
X = X.drop("Price", axis=1)
```

```
X
```

```
y = data[["Price"]]
```

```
y
```

```
from sklearn.linear_model import LinearRegression
```

```
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.10)
```

```
X_train
```

```
y_train
```

```
model = LinearRegression()
```

```
model.fit(X_train, y_train)
```

```
predictions = model.predict(X_test)
```

```
predictions
```

```
predictions - y_test
```

```
print(mean_squared_error(y_test, predictions))
```
