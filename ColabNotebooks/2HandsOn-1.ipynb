```
def calculator(operand1,operand2,operator):
  if operator == '+':
    return operand1+operand2
  elif operator == '-':
    return operand1-operand2
  elif operator == '*':
    return operand1*operand2
  elif operator == '/':
    if operand2 != 0:
      return operand1/operand2
    else:
      return None
  else:
    print("invalid inputs")
    return None
  
```

```
print(calculator(2,3,'_'))
```
